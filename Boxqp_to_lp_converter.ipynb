```python
from glob import glob as glb
from os import path
from gurobipy import *

fsplit = path.splitext
```

```python
### User input: file path stuff ###

p_classes = ['basic', 'small', 'extended', 'extended2']
p_classes = ['basic']

for pcls in p_classes:
    myfpath = fr'./boxQP/data/{pcls}/*.in'
    fpaths = glb(myfpath)
    for fpath in fpaths:
        #fpath = 'lolcat.in'
        fpath = './boxQP/data/basic\spar020-100-1_v2.in'
        print(fpath)
        (fp, fnameext) = path.split(fpath)
        (pname, fext) = fsplit(fnameext)
        
        fin_ext = '.in'
        fout_ext = '.lp'
        
        #### Open files ###
        fnin = fpath
        fnout = fr'./boxQP/LP-data/{pcls}-LP/{pname}.lp'

        fin = open(fnin, 'r')
        fout = open(fnout, 'w')

        ### Read in data ###
        Lin = [[float(s) for s in l.strip().split(' ')] for l in fin.readlines()]

        n = int(Lin[0][0])
        c = Lin[1]
        Q = Lin[2:]

        #n=2
        #c = [1,0]
        #Q = [[-4,2],[2,-2]]
    
        I = range(n)

        # Symmetrize Q, store in lower triangle 
        Qissym = all(Q[i][j] == Q[j][i] for i in I for j in I if i>j)
        if not Qissym:
            for i in I:
                for j in I:
                    if i>j:
                        Q[i][j] = 0.5*(Q[i][j]+Q[j][i])

        ### Define helper functions ###
        pout = lambda s : print(s, file=fout)

        def sign_str(c:float, isfirst = True):
            plus_str = ('' if isfirst else ' + ')
            minus_str = ' - '
            sstr = '' if s == 0 else plus_str if s>0 else minus_str
            
            return f'{sstr}{abs(c)}'
        
        def qstr(i,j,isfirst = True):
            # Q is symmetric, so double bilinear coefficients
            
            c = Q[i][j] * (0.5 if i==j else 1)
            cstr = sign_str(c, isfirst)
            
            if i==j:
                return f'{cstr} x{i} ^ 2'
            else:
                return f'{cstr} x{i} * x{j}'

        ### Write data ###

        ## Objective
        pout('Minimize')

        cstr = ''.join(f'{signstr(c[i], i)} x{i}' 
                          for i in I if c[i]!=0)
        
        if len(cstr) >= 1:
            cstr += ' + '
        
        Q00str = ('',qstr(0,0,False))[Q[0][0]!=0]
        # Q is symmetric, so only write lower triangle. Bilinear coefficients are doubled for correct conversion.
        Qstr = Q00str + ''.join(qstr(i,j) 
                                for i in I for j in I 
                                if i+j>=1 if i>=j if Q[i][j] != 0)

        
        pout('  ' + cstr + Qstr)

        # Test thing
        #pout('Subject To')
        
        ## Variables bounds
        pout('Bounds')            
        pout('  ' + '\n  '.join(f'0 <= x{i} <= 1' 
                                for i in I)) 

        #pout('Generals')
        ## List of variable names
        #pout('  ' + ' '.join(f'x{i}' for i in I))
        #pout('End')        

        fout.close()
        fin.close()
        #raise ValueError('bananas')
        
```

```python
### User input: file path stuff ###

p_classes = ['basic', 'small', 'extended', 'extended2']

for pcls in p_classes:
    myfpath = fr'./boxQP/data/{pcls}/*.in'
    fpaths = glb(myfpath)
    for fpath in fpaths:
        #fpath = 'lolcat.in'
        #fpath = './boxQP/data/basic\spar020-100-1_v2.in'
        print(fpath)
        (fp, fnameext) = path.split(fpath)
        (pname, fext) = fsplit(fnameext)
        
        fin_ext = '.in'
        fout_ext = '.lp'
        
        #### Open files ###
        fnin = fpath
        fnout = fr'./boxQP/LP-data/all-LP/{pname}.lp'

        fin = open(fnin, 'r')

        ### Read in data ###
        Lin = [[float(s) for s in l.strip().split(' ')] for l in fin.readlines()]

        n = int(Lin[0][0])
        c = Lin[1]
        Q = Lin[2:]
    
        I = range(n)

        ### Write data directly via Gurobi###

        ## Objective
        mdl = Model()
        
        x = mdl.addVars(I, name='x', lb = 0, ub = 1)
        mdl.setObjective(quicksum(-c[i]*x[i] for i in I)
                         + quicksum(-0.5*Q[i][j] *x[i]*x[j] for i in I for j in I 
                                    if Q[i][j] != 0))
        
        mdl.update()
        mdl.write(fnout)
        
```

```python
# Solve all the things
from gurobipy import *
import pprint
pp = pprint.PrettyPrinter(indent=2)

times = []
p_classes = ['basic', 'small', 'extended', 'extended2']
for pcls in p_classes:
    # file path per pcls; user input
    myfpath = fr'./boxQP/LP-data/{pcls}-LP/*.lp'
    fpaths = glb(myfpath)
    for fpath in fpaths:
        (fp, fnameext) = path.split(fpath)
        (pname, fext) = fsplit(fnameext)
        
        m = read(fpath)

        # Set all variables to continuous, to circumvent something weird with Gurobi reading in variables as integers
        #   if there is no 'Subject To' line in the .lp file.
        vs = m.getVars()
        for i in range(len(vs)):
            m.setAttr("vtype", [vs[i]], [GRB.CONTINUOUS])

        m.setParam('NonConvex', 2)
        m.setParam('TimeLimit', 20)
        m.optimize()

        times.append((fp+'/'+pname, m.getAttr(GRB.Attr.Runtime)))
                     
pp.pprint(times)
```
